```python
"""
Handwritten Character Recognition
Task 3: Code Alpha ML Internship
Dataset: MNIST (Digits) and EMNIST (Characters)
Model: Convolutional Neural Network (CNN)
"""

# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

print("="*70)
print("HANDWRITTEN CHARACTER RECOGNITION USING CNN")
print("Dataset: MNIST (Handwritten Digits)")
print("="*70)

# ==================== 1. LOAD AND EXPLORE DATASET ====================
print("\n1. Loading MNIST Dataset...")

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"Training samples: {X_train.shape[0]}")
print(f"Testing samples: {X_test.shape[0]}")
print(f"Image shape: {X_train.shape[1:]} (28x28 pixels)")
print(f"Number of classes: {len(np.unique(y_train))} (digits 0-9)")

# Display class distribution
print(f"\nClass Distribution in Training Set:")
unique, counts = np.unique(y_train, return_counts=True)
for digit, count in zip(unique, counts):
    print(f"  Digit {digit}: {count} samples")

# ==================== SAVE ORIGINAL DATASET ====================
print("\n Saving Dataset Files...")

# Save training data as CSV (flattened 28x28 images into 784 columns)
print("  Preparing training dataset...")
train_df = pd.DataFrame(X_train.reshape(X_train.shape[0], -1))
train_df.columns = [f'pixel_{i}' for i in range(784)]  # Name columns pixel_0 to pixel_783
train_df['label'] = y_train
train_df.to_csv('mnist_train_dataset.csv', index=False)
print(f"   Saved: mnist_train_dataset.csv ({X_train.shape[0]} samples, 785 columns)")

# Save test data as CSV
print("  Preparing test dataset...")
test_df = pd.DataFrame(X_test.reshape(X_test.shape[0], -1))
test_df.columns = [f'pixel_{i}' for i in range(784)]
test_df['label'] = y_test
test_df.to_csv('mnist_test_dataset.csv', index=False)
print(f"   Saved: mnist_test_dataset.csv ({X_test.shape[0]} samples, 785 columns)")

# Save a small sample for quick viewing
sample_df = train_df.head(100)
sample_df.to_csv('mnist_sample_100.csv', index=False)
print(f"   Saved: mnist_sample_100.csv (100 samples for quick preview)")

print("\n Dataset Structure:")
print(f"   Each row = one 28x28 image")
print(f"   Columns 0-783 = pixel values (0-255)")
print(f"   Last column = digit label (0-9)")

# ==================== 2. VISUALIZE SAMPLE IMAGES ====================
print("\n2. Visualizing Sample Images...")

# Display random samples from each digit
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle('Sample Handwritten Digits from MNIST', fontsize=16, fontweight='bold')

for i in range(10):
    ax = axes[i // 5, i % 5]
    idx = np.where(y_train == i)[0][0]
    ax.imshow(X_train[idx], cmap='gray')
    ax.set_title(f'Digit: {i}', fontsize=12, fontweight='bold')
    ax.axis('off')

plt.tight_layout()
plt.savefig('sample_digits.png', dpi=300, bbox_inches='tight')
print(" Saved: sample_digits.png")
plt.show()

# ==================== 3. DATA PREPROCESSING ====================
print("\n3. Preprocessing Data...")

# Normalize pixel values to [0, 1]
X_train_normalized = X_train.astype('float32') / 255.0
X_test_normalized = X_test.astype('float32') / 255.0

# Reshape for CNN input
X_train_reshaped = X_train_normalized.reshape(-1, 28, 28, 1)
X_test_reshaped = X_test_normalized.reshape(-1, 28, 28, 1)

# One-hot encode labels
y_train_encoded = to_categorical(y_train, 10)
y_test_encoded = to_categorical(y_test, 10)

print(f"Training data shape: {X_train_reshaped.shape}")
print(" Preprocessing completed")

# ==================== 4. BUILD CNN MODEL ====================
print("\n4. Building CNN Model...")

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),
    
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),
    
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n Model Architecture:")
model.summary()

# ==================== 5. TRAIN THE MODEL ====================
print("\n5. Training the Model...")

history = model.fit(
    X_train_reshaped, y_train_encoded,
    epochs=15,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

print("\n Model training completed!")

# ==================== 6. EVALUATE MODEL ====================
print("\n6. Evaluating Model Performance...")

test_loss, test_accuracy = model.evaluate(X_test_reshaped, y_test_encoded, verbose=0)
print(f"\nTest Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

y_pred_probs = model.predict(X_test_reshaped, verbose=0)
y_pred = np.argmax(y_pred_probs, axis=1)

# ==================== 7. VISUALIZE TRAINING HISTORY ====================
print("\n7. Visualizing Training History...")

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

axes[0].plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
axes[0].plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(alpha=0.3)

axes[1].plot(history.history['loss'], label='Training Loss', linewidth=2)
axes[1].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
print(" Saved: training_history.png")
plt.show()

# ==================== 8. CONFUSION MATRIX ====================
print("\n8. Creating Confusion Matrix...")

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.title('Confusion Matrix', fontsize=16, fontweight='bold')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
print(" Saved: confusion_matrix.png")
plt.show()

# ==================== 9. CLASSIFICATION REPORT ====================
print("\n" + "="*70)
print("CLASSIFICATION REPORT")
print("="*70)
print(classification_report(y_test, y_pred, target_names=[f'Digit {i}' for i in range(10)]))
print("="*70)

# Save classification report to file
with open('classification_report.txt', 'w') as f:
    f.write("="*70 + "\n")
    f.write("HANDWRITTEN DIGIT RECOGNITION - CLASSIFICATION REPORT\n")
    f.write("="*70 + "\n\n")
    f.write(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)\n")
    f.write(f"Test Loss: {test_loss:.4f}\n\n")
    f.write(classification_report(y_test, y_pred, target_names=[f'Digit {i}' for i in range(10)]))

print(" Saved: classification_report.txt")

# ==================== 10. VISUALIZE PREDICTIONS ====================
print("\n10. Visualizing Sample Predictions...")

# Show some correct and incorrect predictions
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle('Sample Predictions (Green=Correct, Red=Incorrect)', fontsize=16, fontweight='bold')

for i in range(10):
    ax = axes[i // 5, i % 5]
    idx = np.random.randint(0, len(X_test))
    
    ax.imshow(X_test[idx], cmap='gray')
    true_label = y_test[idx]
    pred_label = y_pred[idx]
    
    color = 'green' if true_label == pred_label else 'red'
    ax.set_title(f'True: {true_label} | Pred: {pred_label}', 
                fontsize=11, fontweight='bold', color=color)
    ax.axis('off')

plt.tight_layout()
plt.savefig('prediction_samples.png', dpi=300, bbox_inches='tight')
print(" Saved: prediction_samples.png")
plt.show()

# ==================== 11. SAVE MODEL ====================
print("\n11. Saving Trained Model...")

model.save('handwriting_recognition_model.h5')
print(" Model saved as 'handwriting_recognition_model.h5'")

# Save model architecture as JSON
model_json = model.to_json()
with open('model_architecture.json', 'w') as f:
    f.write(model_json)
print(" Model architecture saved as 'model_architecture.json'")

# ==================== 12. SAVE TRAINING HISTORY ====================
history_df = pd.DataFrame(history.history)
history_df.to_csv('training_history.csv', index=False)
print(" Training history saved as 'training_history.csv'")

# ==================== SUMMARY ====================
print("\n" + "="*70)
print(" PROJECT COMPLETED SUCCESSFULLY!")
print("="*70)
```
